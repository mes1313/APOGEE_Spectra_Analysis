```python
from astropy.table import join
apogee_data_table = Table.from_pandas(df)

zeros = [0] * len(apogee_data_table)

def createList(r1, r2):
    return list(range(r1, r2))

r1, r2 = 0, len(apogee_data_table)
indices = createList(r1, r2)
apogee_data_table.add_column(zeros, index = len(apogee_data_table.columns), name = "Radius")
apogee_data_table.add_column(indices, index = 0, name = "Index")

print(apogee_data_table)
```
